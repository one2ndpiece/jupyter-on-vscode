```python
from moviepy.editor import VideoFileClip
import matplotlib.pyplot as plt

# 入力動画の読み込み
clip = VideoFileClip(r"C:\Users\toran\Videos\Captures\クエリエディタ _ Athena _ ap-southeast-2 - Google Chrome 2023-09-22 15-20-35.mp4")

# 抽出するフレームの時間を指定（秒）
time = 5.0  # 5秒目のフレームを抽出

# フレームを抽出
frame = clip.get_frame(time)

# トリミングする座標を指定（x1, y1, x2, y2）
x1, y1, x2, y2 = 0, 100, 2000, 2000

# トリミング処理
trimmed_frame = frame[y1:y2, x1:x2]

# トリミング後のフレームを表示
plt.imshow(trimmed_frame)
plt.axis('off')
plt.show()
```
