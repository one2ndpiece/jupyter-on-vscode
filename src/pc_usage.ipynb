```python
import psutil
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# Initial setup
fig, ax = plt.subplots(2)
cpu_data = []
memory_data = []

# Set the title and labels
ax[0].set_title('CPU Usage')
ax[1].set_title('Memory Usage')
ax[0].set_ylim([0, 100])
ax[1].set_ylim([0, 100])

while True:
    # Get the current CPU and memory usage
    cpu_usage = psutil.cpu_percent(interval=1)
    memory_usage = psutil.virtual_memory().percent

    # Append the usage data
    cpu_data.append(cpu_usage)
    memory_data.append(memory_usage)

    # If the data lists are too long, remove the oldest data point
    if len(cpu_data) > 50:
        del cpu_data[0]
        del memory_data[0]

    # Update the plots
    ax[0].cla()
    ax[1].cla()
    ax[0].plot(cpu_data, label=f'CPU Usage: {cpu_usage}%')
    ax[1].plot(memory_data, label=f'Memory Usage: {memory_usage}%')
    ax[0].legend(loc='upper right')
    ax[1].legend(loc='upper right')

    # Clear the previous output and display the updated plot
    clear_output(wait=True)
    display(fig)
    
    # Pause for a brief moment to allow the plot to update
    plt.pause(0.1)
```
